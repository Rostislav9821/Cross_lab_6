# Задача:
Тело массой m = 1 кг брошено c поверхности земли вертикально вверх с начальной скоростью U_{0} = 19,6 м/с. Определить изменение потенциальной энергии тела за промежуток времени t = 2 с после броска. Построить графики зависимости Uy(t) и Еp(t). Сопротивление воздуха не учитывать.

Скорость U тела изменяется так:

U=U0-g*t

Нетрудно заметить, что по истечении двух секунд тело будет находиться в точке наивысшего подъема, поэтому

h=U0^2/2*g

А потенциальная энергия изменится на

Ep=m*g*h=m*U0^2/2=19,6^2/2=192

Ответ: энергия изменится на 192 Дж.

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from matplotlib import cm
import matplotlib.pyplot as plt

f = lambda x, y: x ** 2 - y ** 2

fig = plt.figure(figsize = (10, 10))

ax = fig.add_subplot(1, 1, 1, projection = '3d')

xval = np.linspace(-20, 20, 200)
yval = np.linspace(-20, 20,200)

x, y = np.meshgrid(xval, yval)

z = f(x, y)

surf = ax.plot_surface(

x, y, z, 

rstride = 2,
cstride = 2,

cmap = cm.viridis)
```
